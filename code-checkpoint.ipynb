```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load the dataset
data = pd.read_csv('employee_attrition.csv') 

# View basic info about the dataset
print(data.info())
```

```python
# Checking for missing values
print(data.isnull().sum())


# Fill missing job satisfaction scores with the median value
data['JobSatisfaction'] = data['JobSatisfaction'].fillna(data['JobSatisfaction'].median())
```

```python
# Convert 'Attrition' column from Yes/No to binary (1/0)
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Plot attrition distribution
sns.countplot(data['Attrition'])
plt.title('Attrition Distribution')
plt.show()
```

```python
# Convert categorical variables to numeric using one-hot encoding
data_encoded = pd.get_dummies(data, drop_first=True)  # drop_first=True avoids multicollinearity by dropping one category

# Now generate the correlation matrix on the encoded data
plt.figure(figsize=(12, 8))
sns.heatmap(data_encoded.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Plot relationships between attrition and key variables (e.g., MonthlyIncome, JobSatisfaction)
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title('Attrition vs Monthly Income')
plt.show()

sns.boxplot(x='Attrition', y='JobSatisfaction', data=data)
plt.title('Attrition vs Job Satisfaction')
plt.show()

# Plotting attrition rate by department
sns.countplot(x='Department', hue='Attrition', data=data)
plt.title('Attrition by Department')
plt.show()
```

```python
# Selecting relevant features for modeling
features = ['Age', 'Department', 'JobRole', 'MonthlyIncome', 'JobSatisfaction', 'YearsAtCompany']
X = pd.get_dummies(data[features], drop_first=True)  # Use one-hot encoding for categorical variables
y = data['Attrition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict on the test data
y_pred = clf.predict(X_test)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Feature importance
importance = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=importance, y=importance.index)
plt.title('Feature Importance for Attrition Prediction')
plt.show()
```
